```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python

d1=pd.read_csv('monthly-hpi.csv') #Monthly Housing Price Index

d2=pd.read_csv('fed_funds.csv') #Federal Funds

d3=pd.read_csv('gdp.csv') #GDP

d4=pd.read_csv('shiller.csv') #Shiller

d5=pd.read_csv('unemployment-macro.csv') #Unemployment CSV
```

```python
data=pd.merge(d1, d2, on='date')
data=pd.merge(data, d3, on='date')
data=pd.merge(data, d4, on='date')
data=pd.merge(data, d5, on='date')

#data made with first column as date to match all the data

data=data.iloc[:,1:]

#data refined after removing date as a criteria

y=data.iloc[:,0] # Y is HPI
y=y.astype('int')
data=data.iloc[:,1:]
#print y
print(data.columns)
```

### We now use RFE -> Recursive Feature Elimination which uses recursive cross-validation to determine which features are important. 

### After repeated tries, we found that the best accuracy is delivered when we use 7 features.

```python
model=linear_model.LogisticRegression()

from sklearn.feature_selection import RFE
rfe=RFE(model,7)
rfe=rfe.fit(data,y)

print(rfe.support_)
print(rfe.ranking_)
a=np.array([0,0,0,0,0,0,0])

j=0

print(len(data.columns))

print(rfe.ranking_[13])

for i in range(0,len(data.columns)):
    if(rfe.ranking_[i]==1):
        a[j]=i
        #print(j,'->')
        j+=1

#print a
        
data1=data
 
#Trimming the input attributes to those attributes which are accepted by RFE

data=data.iloc[:,a]

print(data)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size = 0.2, random_state = 0)
```

### Used Support Vector Machine, Gaussian and Multinomial Naive Bayes, but Linear Regression turned out to give the best results for the given data.

```python
#from sklearn.svm import SVC	
#classifier = SVC()
#classifier.fit(X_train,y_train)

# Fitting Naive Bayes to the Training set
#from sklearn.naive_bayes import GaussianNB
#classifier=GaussianNB()
#classifier.fit(X_train,y_train)

#Linear Regression


classifier=linear_model.LinearRegression()
classifier.fit(X_train,y_train)

filename = 'finalized_model.sav'
pickle.dump(classifier, open(filename, 'wb'))
```

```python

# Predicting the Test set results
y_pred = classifier.predict(X_test)
print("Actual test values")
print(y_test)

print("Predicted Values")
print(y_pred)
#print(type(y_test))
#print(type(X_test))
```

### Mean Absolute Error

### Variance Regression Score: 1 is favorable

```python
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_absolute_error

meanscore=mean_absolute_error(y_test,y_pred,multioutput='uniform_average')
varscore=explained_variance_score(y_test, y_pred)
print('The mean absolute error is ',meanscore)
print('The variance score is ', varscore)
```

### According to the trained model, plotting graph of each variable against the HPI 

```python
#plot outputs
#x1=X_test['gross_domestic_product']

col=len(X_test.columns)
for i in range(0, col):
	plt.scatter(X_test.iloc[:,i], y_test,  color='black')
	plt.plot(X_test.iloc[:,i], y_pred, color='blue', linewidth=3)
	plt.xlabel(X_test.columns.get_values()[i])
	plt.ylabel('HPI')	
	plt.show()



#ax = x1.plot()
#y_test.plot(ax=ax)
```
